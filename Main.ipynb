### Analyzing Data Distributions

Let's inspect the distributions of numerical features for ranges, skewness, and potential outliers using histograms and box plots. For categorical features, we'll examine value counts and their distributions.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Identify numerical and categorical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object', 'bool']).columns

print("\n--- Numerical Feature Distributions ---")
for col in numerical_cols:
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    # Histogram
    sns.histplot(df[col], kde=True, ax=axes[0])
    axes[0].set_title(f'Distribution of {col}')
    axes[0].set_xlabel(col)
    axes[0].set_ylabel('Frequency')

    # Box Plot
    sns.boxplot(y=df[col], ax=axes[1])
    axes[1].set_title(f'Box Plot of {col}')
    axes[1].set_ylabel(col)
    plt.tight_layout()
    plt.show()
```

```python
print("\n--- Categorical Feature Distributions ---")
for col in categorical_cols:
    print(f"\nValue counts for '{col}':")
    display(df[col].value_counts())

    # Optional: Plotting top N categories if too many unique values
    if df[col].nunique() < 20: # Limit to avoid overly dense plots
        plt.figure(figsize=(10, 6))
        sns.countplot(y=df[col], order=df[col].value_counts().index, palette='viridis')
        plt.title(f'Count of {col}')
        plt.xlabel('Count')
        plt.ylabel(col)
        plt.tight_layout()
        plt.show()
```

### Checking for Leakage and Redundancy Risks

Leakage occurs when information from the target variable is inadvertently included in the features. Redundancy means having highly correlated features that provide similar information. We'll examine the correlation matrix for numerical features and consider the relationship between features and our target variable, `Price_SGD`.

```python
print("\n--- Correlation Matrix of Numerical Features ---")
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()

# Specifically look at correlations with the target variable
print("\nCorrelation with Price_SGD:")
display(correlation_matrix['Price_SGD'].sort_values(ascending=False))
```

### Addressing Data Leakage and Redundancy

Given the insights from the correlation matrix and the scatter plot:

*   **`Member_Discount` and `Price_SGD`:** The strong, linear relationship observed suggests `Member_Discount` is likely data leakage, a direct derivative of `Price_SGD`. Including it would lead to an overfitted model. We will drop this column.
*   **`Weight_kg` and `Screen_Size_inch`:** These features are highly correlated (0.92), indicating redundancy. To simplify the model and avoid multicollinearity, we will drop `Weight_kg`, as `Screen_Size_inch` often provides more direct information about a laptop's form factor.

```python
print('Dropping Member_Discount due to data leakage and Weight_kg due to high correlation with Screen_Size_inch.')
df = df.drop(['Member_Discount', 'Weight_kg'], axis=1)

print('\nDataFrame after dropping problematic columns:')
display(df.info())

# Update numerical_cols to reflect the dropped columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object', 'bool']).columns
```

### Categorical Feature Encoding

To prepare our categorical features for machine learning models, we need to convert them into a numerical format. Given the nature of these features (many unique values for 'Model', 'CPU', 'GPU', and 'Brand'), one-hot encoding is a suitable approach for most, while a simple binary encoding will work for 'Touchscreen'.

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# Separating features for encoding
categorical_cols = df.select_dtypes(include=['object', 'bool']).columns

# One-Hot Encoding for multi-category nominal features
one_hot_cols = [col for col in categorical_cols if col != 'Touchscreen']

# Initialize OneHotEncoder
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# Apply one-hot encoding
encoded_features = encoder.fit_transform(df[one_hot_cols])
encoded_feature_names = encoder.get_feature_names_out(one_hot_cols)

df_encoded = pd.DataFrame(encoded_features, columns=encoded_feature_names, index=df.index)

# Handle 'Touchscreen' separately as it's a boolean (binary)
# Convert boolean to int (True=1, False=0)
df_encoded['Touchscreen'] = df['Touchscreen'].astype(int)

# Drop original categorical columns and concatenate encoded features
df_processed = pd.concat([df.drop(columns=categorical_cols), df_encoded], axis=1)

print("DataFrame after One-Hot Encoding and Touchscreen conversion:")
display(df_processed.head())
display(df_processed.info())
```
